```python
# 확인할 디렉토리 경로 설정
def dir_files_num(path):
    directory = path

    # 디렉토리 내의 파일 목록 가져오기
    file_list = os.listdir(directory)

    # 파일 개수 출력
    print(f"디렉토리 '{directory}'에 있는 파일 수: {len(file_list)}")

dir_files_num('./data/labels')
dir_files_num('./data/points')
```

```python
# 데이터셋 폴더 경로
data_folder = './data/points'
label_folder = './data/labels'

# 전체 데이터를 저장할 리스트
# all_points = []
# all_labels = []

# 파일 리스트 불러오기 (train 데이터셋 전부)
data_list = os.listdir(data_folder)
data_list = [f[:-3] for f in data_list if f.endswith('.npy')]

label_list = os.listdir(label_folder)
label_list = [f[:-3] for f in label_list if f.endswith('.txt')]

test_label = []
for i in data_list:
    if i not in label_list:
        test_label.append(i + 'npy')
```
